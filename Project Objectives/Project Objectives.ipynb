## Introduction 

Gerrymandering occurs when district boundaries are drawn so that the party creating the districts have a political advantage over other parties. Gerrymandering got its name from the governer of Massachusetts Elbridge Gerry who signed a bill in 1812 that created an unusual districting of the state shown here. This gave the Democrat-Republicans a bigger state Senate majority, even though the Federalist party got more votes statewide.

The Voting Rights Act of 1965 made it illegal to gerrymander by race, i.e. people of color cannot be spread out to dilute their vote. Instead districts may be drawn to group together people with similar interests and form voting blocks. 

Two main ways to gerrymander a map are by packing and cracking. `Packing` means that one political party is packed into a district. `Cracking` dilutes a party's votes by spreading out the voters so that they can't win as many districts.

## Measurements of Gerrymandering

How does one prove that Gerrymandering has occured?
Some ways to measure Gerrymandering are:

- `Efficiency Gap`, difference between the surplus votes for each party divided by the total number of votes. 
    - `Surplus` is given by: 
    (for the winner of the district) the number of votes recieved minus the min amount of votes needed to win a district or 
    (for the loser) the number of votes recieved. 
    - Compare efficiency gap with the percentage each district represents in the state to get a sense of how many more seats the winning party has beyond what would be a proportional representation.
