# Reading shapefiles for Indiana

### Packages and timer function

```python
import pandas as pd
import time
# Input is a start time and an end time
# Output is a print statement giving the time elapsed
def time_elapsed(start, end):
    # Compute time elapsed in seconds
    total_time_seconds = end-start 
    if total_time_seconds < 60:
        print("Total time elapsed =",total_time_seconds, "seconds")
    else:
        # In minutes 
        total_time_minutes = total_time_seconds/60 
        if total_time_minutes < 60: 
            print("Total time elapsed =", total_time_minutes, "minutes") 
        else: 
            # In hours
            total_time_hours = total_time_minutes/60 
            # Print the time elapsed in hours
            print("Total time elapsed =", total_time_hours, "hours")
from dbfread import DBF # for reading dbf files
import geopandas as gpd # for reading shp files
```

### Reading the dbf file from MGGG

```python
start = time.time()
IndianaDBF = pd.DataFrame()
for record in DBF("data/Indiana/MGGG/Indiana.dbf"):
    df = pd.DataFrame([record.values()], columns=record)
    #print(df) # for debugging
    IndianaDBF = pd.concat([IndianaDBF, df], axis=0)
finish = time.time()
time_elapsed(start, finish)
IndianaDBF
```

### Reading the shp file from MGGG

```python
start = time.time()
IndianaSHP = gpd.read_file("data/Indiana/MGGG/Indiana.shp")
finish = time.time()
time_elapsed(start, finish)
IndianaSHP.info()
```

### Shapefile from redistricting.lls.edu

```python
start = time.time()
IndianaSHP2 = gpd.read_file("data/Indiana/AllAboutRedistricting/Congress_Districts_113th_USCB_IN.shp")
finish = time.time()
time_elapsed(start, finish)
IndianaSHP2
```

### Shapefile from redistrictingdatahub.org

```python
start = time.time()
IndianaSHP3 = gpd.read_file("data/Indiana/RedistrictingDataHub/2021_adopted_congress_plan/Congressional2021AdoptedOCT012021.shp")
finish = time.time()
time_elapsed(start, finish)
IndianaSHP3
```
